```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.special import gamma
from scipy.optimize import minimize
```

```python
from mpl_toolkits.mplot3d import Axes3D

#function
f = lambda x1, x2: 6*x1**2 + 3*x2**2 + 4*x1*x2

#Taking the values od x1, x2
x2 = np.linspace(-5,5,100)
x1 = np.linspace(-5,5,100)

X, Y = np.meshgrid(x1,x2)
F = f(X,Y)

# creating the 3D surface plot
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(1,2,1,projection='3d') # set up the axis as a 3D plot

#plotting the solution on the grid
ax.scatter3D(1,1,f(1,1),c="blacl",marker="*",s=100)
surf = ax.plot_surface(X,Y,F, cmap = 'gist_rainbow' )

#Representing the colrbar
# fig.colorbar(surf,shrink=0.4, aspect=10)

# set axes label
ax.set_xlabel('x1', labelpad=10)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)
ax.set_title("3D plot of my function")

plt.show()
```

```python
# fixing the size of the contour plot
plt.figure(figsize=(8,6))

# levels parameter indicates no. of contour lines
# contour function generates the contour plot
plt.contour(X, Y, F, levels = 100, cmap = 'gist_rainbow')

#Adding extra features to the plot for clear understanding
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour plot of my function')
plt.colorbar()
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize

#function
f = lambda x: 6*x[0]**2 + 3*x[1]**2 + 4*x[0]*x[1]

#constraints x0*x1 = 7
constraint = {'type': 'eq', 'fun': lambda x: x[0]*x[1] - 7}

#Initial guess for the minimum
x0 = [0,0]

result = minimize(f,x0, method='SLSQP',constraints = constraint)
print("Minimum value found:", result.fun)
print("Location of minimum:", result.x)
```
