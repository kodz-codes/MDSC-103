# Using Gradiant Descend Method to find out the maximum or minimum

# Importing the required libraries

```python
# importing the libraries
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#defining the parameter for the plot
start = int(input("Enter the start range of values : "))
end = int(input("Enter the end range of x values : "))

x_values = np.arange(start,end,0.1)

x_y = np.array([x**2 for x in x_values])

#Adding title, x, y axis
plt.xlabel("X")
plt.ylabel("f(y)")
plt.title("function : x**2 ")

plt.plot(x_values,x_y)  # plotting the x, y curve

#ALL initializations for the initial values
x0 = 4
x_new = 34
eta = 0.1
diff = 1
count = 0
grad_x = 2*x0

# The Main loop which performs the gradiant descend method 
while(count < 100):
    fx = x0 ** 2
    grad_x = 2 * x0
    x_new = x0 - (eta * grad_x)
    plt.scatter(x0,fx)
    if(abs(x_new-x0) < 0.00001):
        break
    x0 = x_new
    count = count + 1
    
    
print(count,x_new)
```
